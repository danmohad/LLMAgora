# LLM Agora Demo

This notebook is intentionally thin: all core logic lives in `agora.experiment`.
The main runtime options are set explicitly below for ease of learning. The default example runs online and saves a `debate_snapshot.json` when `save_snapshot` is enabled.


```python
from dotenv import load_dotenv

from agora.experiment import build_experiment_config, run_persona_experiment

load_dotenv()
```

```python
config = build_experiment_config(
    {
        # Scenario selection
        # scenario_id: any scenario ID from data/scenarios.json
        "scenario_id": "promotion_committee",
        # incentive_direction=None disables incentive context; otherwise
        # incentive_type selects which module under that direction to use.
        "incentive_direction": 'positive',  # positive | negative | None
        "incentive_type": "future",  # historical | future
        # Prompt template set from data/prompts.json.
        "prompt_set": "default",

        # Model and run length
        # Use any OpenRouter model identifier available to your account.
        "model": "openai/gpt-5.4",
        # Number of debate rounds after optional pre-interviews.
        "num_turns": 2,
        # subturn_event_order drives event enablement directly
        # always include public_utterance
        # include private_utterance to enable private reflection
        # include public_survey to enable public survey
        # include private_survey to enable private survey
        # The list order is the execution order within each agent turn.
        "subturn_event_order": [
            "private_survey",
            "private_utterance",
            "public_survey",
            "public_utterance",
        ],
        # True prints progress and transcript details while the run executes.
        "verbose": True,

        # Private reflection retention. The private_utterance event controls
        # whether reflections run; this flag only controls local memory use.
        "keep_private_reflection": False,

        # Pre/post interviews (always run; keep controls local retention)
        "keep_pre_interview": False,
        "keep_post_interview": False,

        # Survey retention. public_survey/private_survey events control
        # whether surveys run; these flags only control local memory use.
        "keep_public_survey": False,
        "keep_private_survey": False,

        # Semantic similarity analysis
        # options: self_consistency, cross_agent_public_alignment, cross_agent_private_alignment
        # Leave this empty to skip local embedding/NLI analysis.
        "semantic_analysis_metrics": [],
        "semantic_similarity_method": None,  # required when semantic_analysis_metrics is non-empty
        "semantic_similarity_model": None,  # required when semantic_analysis_metrics is non-empty
        "semantic_similarity_device": None,  # None | cpu | mps | cuda

        # Persona adherence analysis
        # options: public_per_turn, private_per_turn, public_cumulative,
        #          private_cumulative, full_debate_public, full_debate_private
        # Leave this empty to skip persona scoring calls.
        "persona_analysis_metrics": [],
        "persona_scoring_model": None,  # required when persona_analysis_metrics is non-empty
        "persona_scoring_verbose": False,
        "persona_score_samples": None,  # required when persona_analysis_metrics is non-empty

        # Plotting
        "save_plots": False,
        "show_plots": False,  # requires save_plots=True

        # Snapshot handling. load_snapshot reads load_dir/debate_snapshot.json;
        # save_snapshot writes debate_snapshot.json in this run's output folder.
        "load_snapshot": False,
        "load_dir": None,  # required when load_snapshot=True
        "save_snapshot": True,

        # Output layout. These paths assume the notebook working directory is
        # notebooks/; adjust to outputs and data/... if running from repo root.
        "outputs_root": "../outputs",
        "run_name": None,  # set a string to force a custom dir name
        "indexed_output": True,  # True -> outputs/<run_id> + index CSV
        "index_csv": None,  # defaults to outputs_root/index.csv

        # Input files
        "catalog_path": "../data/scenarios.json",
        "prompts_path": "../data/prompts.json",
    }
)

```

```python
result = run_persona_experiment(config)
print("Run directory:", result.run_dir)
```
