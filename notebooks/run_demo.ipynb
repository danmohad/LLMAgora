# LLM Agora Demo

```python
from dotenv import load_dotenv

from agora.experiment import build_experiment_config, run_persona_experiment

load_dotenv()
```

```python
config = build_experiment_config(
    {
        # Any scenario ID from data/scenarios.json.
        "scenario_id": "promotion_committee",
        # None disables incentive context; otherwise choose positive or negative.
        "incentive_direction": "positive",
        # Selects the historical or future module under the chosen incentive direction.
        "incentive_type": "future",
        # Prompt template set from data/prompts.json.
        "prompt_set": "default",

        # Any OpenRouter model identifier available to your account.
        "model": "openai/gpt-5.4",
        # Number of debate rounds after optional pre-interviews.
        "num_turns": 2,
        # Event order within each agent turn; include public_utterance and optional private/survey events.
        "subturn_event_order": [
            "private_survey",
            "private_utterance",
            "public_survey",
            "public_utterance",
        ],
        # True prints progress and transcript details while the run executes.
        "verbose": True,

        # Retention only; private_utterance in subturn_event_order controls whether reflections run.
        "keep_private_reflection": False,
        # Retention only; pre-interviews run before debate turns.
        "keep_pre_interview": False,
        # Retention only; post-interviews run after debate turns.
        "keep_post_interview": False,
        # Retention only; public_survey in subturn_event_order controls whether public surveys run.
        "keep_public_survey": False,
        # Retention only; private_survey in subturn_event_order controls whether private surveys run.
        "keep_private_survey": False,

        # Semantic metrics to run; [] skips local embedding/NLI analysis.
        "semantic_analysis_metrics": [],
        # Required when semantic_analysis_metrics is non-empty.
        "semantic_similarity_method": None,
        # Required when semantic_analysis_metrics is non-empty.
        "semantic_similarity_model": None,
        # Semantic model device: None, cpu, mps, or cuda.
        "semantic_similarity_device": None,

        # Persona adherence metrics to run; [] skips persona scoring calls.
        "persona_analysis_metrics": [],
        # Required when persona_analysis_metrics is non-empty.
        "persona_scoring_model": None,
        # True prints detailed persona scoring progress.
        "persona_scoring_verbose": False,
        # Required when persona_analysis_metrics is non-empty.
        "persona_score_samples": None,

        # Whether to save plots for enabled analyses.
        "save_plots": False,
        # Whether to display plots while running; requires save_plots=True.
        "show_plots": False,

        # Whether to read load_dir/debate_snapshot.json.
        "load_snapshot": False,
        # Required snapshot directory when load_snapshot=True.
        "load_dir": None,
        # Whether to write debate_snapshot.json in this run's output folder.
        "save_snapshot": True,

        # Root output directory, resolved from the notebook working directory.
        "outputs_root": "../outputs",
        # Set a string to force a custom output directory name.
        "run_name": None,
        # True writes outputs_root/<run_id> and appends an index CSV row.
        "indexed_output": True,
        # Optional index CSV path; None defaults to outputs_root/index.csv.
        "index_csv": None,

        # Scenario catalog path, resolved from the notebook working directory.
        "catalog_path": "../data/scenarios.json",
        # Prompt catalog path, resolved from the notebook working directory.
        "prompts_path": "../data/prompts.json",
    }
)

```
