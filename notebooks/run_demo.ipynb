# LLM Agora Demo

This notebook is intentionally thin: all core logic lives in `agora.experiment`.
Every exposed runtime option is set explicitly below for ease of learning.


```python
from dotenv import load_dotenv

from agora.experiment import build_experiment_config, run_persona_experiment

load_dotenv()
```

```python
config = build_experiment_config(
    {
        # Scenario selection
        # scenario_id: any scenario ID from data/scenarios.json
        "scenario_id": "extreme_divergence_credit_theft",
        "question_variant": "controversial",  # agreeable | controversial
        "side_order": "12",  # 12 | 21
        "prompt_set": "default",
        "use_neutral_arena": False,  # True -> use neutral arena prompt instead of persona arena

        # Models and run length
        "alpha_model": "openai/gpt-5.2",
        "beta_model": "anthropic/claude-sonnet-4.5",
        "num_turns": 1,
        # subturn_event_order drives event enablement directly
        # always include public_utterance
        # include private_utterance to enable private reflection
        # include public_survey to enable public survey
        # include private_survey to enable private survey
        "subturn_event_order": [
            "public_utterance",
            "private_utterance",
            "public_survey",
            "private_survey",
        ],
        "verbose": False,

        # Private reflection
        "keep_private_reflection": False,

        # Pre/post interviews
        "enable_pre_interview": True,
        "keep_pre_interview": False,
        "enable_post_interview": True,
        "keep_post_interview": False,

        # Surveys
        "keep_public_survey": False,
        "keep_private_survey": False,

        # Semantic similarity analysis
        # options: self_consistency, cross_agent_public_alignment, cross_agent_private_alignment
        "semantic_analysis_metrics": [
            "self_consistency",
            "cross_agent_public_alignment",
            "cross_agent_private_alignment",
        ],

        # Persona adherence analysis
        # options: public_per_turn, private_per_turn, public_cumulative,
        #          private_cumulative, full_debate_public, full_debate_private
        "persona_analysis_metrics": [],
        "persona_scoring_model": None,  # e.g. anthropic/claude-sonnet-4
        "persona_scoring_verbose": False,
        "persona_score_samples": None, # e.g. 1

        # Plotting
        "save_plots": True,
        "show_plots": False,  # requires save_plots=True

        # Snapshot handling
        "load_snapshot": False,
        "load_dir": None,  # required when load_snapshot=True
        "save_snapshot": True,

        # Output layout
        "outputs_root": "../outputs",
        "run_name": None,  # set a string to force a custom dir name
        "indexed_output": False,  # True -> outputs/<run_id> + index CSV
        "index_csv": None,  # defaults to outputs_root/index.csv

        # Input files
        "catalog_path": "../data/scenarios.json",
        "prompts_path": "../data/prompts.json",
    }
)

```

```python
result = run_persona_experiment(config)
print("Run directory:", result.run_dir)
```
