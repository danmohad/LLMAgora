# Text output (LaTeX PDF)

Edit **`TEXT`** in the cell below, then run it. The notebook renders your text with LaTeX (Computer Modern serif) and saves a PDF directly under `outputs/saved_plots/` (not in a category subfolder).

**Line breaks**
- End a line with `\\` (two backslashes in the string) for a LaTeX line break after that line.
- A line that is only `\\` (or blank) adds extra vertical space between blocks.
- Lines without a trailing `\\` still appear on their own row.

Requires a LaTeX install (MacTeX / TeX Live / MiKTeX). On macOS, the setup cell adds `/Library/TeX/texbin` to `PATH` if needed.

```python
from pathlib import Path
import os
import sys

project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
os.chdir(project_root)
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

import matplotlib as mpl
import matplotlib.pyplot as plt

if "/Library/TeX/texbin" not in os.environ.get("PATH", ""):
    os.environ["PATH"] = "/Library/TeX/texbin:" + os.environ.get("PATH", "")

mpl.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "axes.unicode_minus": False,
})

SAVED_PLOTS_DIR = Path("outputs/saved_plots")
SAVED_PLOTS_DIR.mkdir(parents=True, exist_ok=True)
SAVE_DPI = 300

print("Working directory:", Path.cwd().resolve())
print("PDF output folder:", SAVED_PLOTS_DIR.resolve())
```

```python
# --- Edit TEXT below, then run this cell ------------------------------------

TEXT = r"""a)
b)
c)
d)
e)
f)
\\
Decision stance\\
Cosine self-consistency \\
NLI distribution\\
Survey responses\\
Emotion radars\\
Junior faculty\\
Senior faculty \\
Policy director\\
Coalition chair \\
Mid-level manager\\
Senior Leader
Coalition chair \\
"""

# Saved as outputs/saved_plots/<OUTPUT_STEM>.pdf
OUTPUT_STEM = "text_output"
FONT_SIZE = 14
FIGURE_WIDTH_IN = 5.5


def _tex_escape(s: str) -> str:
    s = str(s)
    for ch in ("_", "&", "%", "$", "#", "{", "}"):
        s = s.replace(ch, "\\" + ch)
    s = s.replace("-", "{-}")
    return s


def _text_to_latex(body: str) -> str:
    """Turn notebook TEXT into a flush-left LaTeX block."""
    chunks: list[str] = []
    for line in body.splitlines():
        raw = line.rstrip()
        if not raw or raw.replace("\\", "").strip() == "":
            chunks.append(r"\vspace{0.55em}")
            continue
        s = raw
        while s.endswith("\\"):
            s = s[:-1].rstrip()
        chunks.append(_tex_escape(s) + r" \\")
    if chunks and chunks[-1].endswith(r" \\"):
        chunks[-1] = chunks[-1][:-4]
    return "\n".join(chunks)


def render_text_pdf(
    text: str,
    out_path: Path,
    *,
    font_size: float = 14,
    width_in: float = 5.5,
    dpi: int = 300,
) -> Path:
    latex = _text_to_latex(text)
    n_lines = max(1, len([ln for ln in text.splitlines() if ln.strip()]))
    height_in = max(2.0, 0.38 * n_lines + 0.8)

    fig = plt.figure(figsize=(width_in, height_in), facecolor="white")
    fig.text(
        0.06,
        0.94,
        latex,
        ha="left",
        va="top",
        fontsize=font_size,
        transform=fig.transFigure,
    )
    out_path = Path(out_path).resolve()
    out_path.parent.mkdir(parents=True, exist_ok=True)
    fig.savefig(out_path, bbox_inches="tight", dpi=dpi, format="pdf", facecolor="white")
    plt.show()
    plt.close(fig)
    return out_path


out_pdf = render_text_pdf(
    TEXT,
    SAVED_PLOTS_DIR / f"{OUTPUT_STEM}.pdf",
    font_size=FONT_SIZE,
    width_in=FIGURE_WIDTH_IN,
    dpi=SAVE_DPI,
)
print(f"Saved: {out_pdf}")
```
